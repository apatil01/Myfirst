```python
import pandas as pd
import numpy as np
import re
```

```python
#data = pd.read_csv("Followers_of_PEX.csv")
#data = pd.read_csv("Followers_of_Square.csv")
data = pd.read_csv(open("Followers_of_Square.csv",'rU'), encoding='utf-8', engine='c')
data.dropna(subset=['screen_name'], how='any', inplace=True)
data['description'] = data['description'].apply(lambda x : str(x) if isinstance(x, float) else str(x.encode('utf-8','ignore')).lower())
```

```python
data['description'] = data['description'].apply(lambda x : re.sub("[^a-zA-Z ]","", str(x)))
data['description'] = data['description'].apply(lambda x : str(x).replace("?",""))
data['description'] = data['description'].apply(lambda x : str(x).replace("!",""))
data['description'] = data['description'].apply(lambda x : str(x).replace("#",""))
data['description'] = data['description'].apply(lambda x : str(x).replace(".",""))
data['description2'] =data['description'].apply(lambda x : str(x).split())
```
